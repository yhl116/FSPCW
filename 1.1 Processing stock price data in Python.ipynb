```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#question1
px = pd.read_csv("priceData.csv")
```

```python
logpx = np.log(px["SPX Index"])
```

```python
px["Log SPX Index"] = logpx
```

```python
plt.figure(figsize=(20,10))
plt.plot(logpx)  
plt.title("Log of SPX Index")  
plt.xlabel("Index")  
plt.ylabel("Log of SPX Index")  
plt.grid()
plt.show()
```

```python
#question2
logpx_rolling_mean = logpx.rolling(252, min_periods=1).mean()
logpx_rolling_std = logpx.rolling(252, min_periods=1).std()
```

```python
plt.figure(figsize=(20,10))
plt.plot(logpx_rolling_mean)  
plt.title("Log of SPX Index with Rolling Mean")  
plt.xlabel("Index")  
plt.ylabel("Log of SPX Index with Rolling Mean")  
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(logpx_rolling_std)  
plt.title("Log of SPX Index with Rolling Standard Deviation")  
plt.xlabel("Index")  
plt.ylabel("Log of SPX Index with Rolling Standard Deviation")  
plt.grid()
plt.show()
```

```python
#question3
logret = logpx.diff()
simpret = px["SPX Index"].pct_change()
```

```python
plt.figure(figsize=(20,10))
plt.plot(logret)  
plt.title("Log Return")  
plt.xlabel("Index")  
plt.ylabel("Log Return")  
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(simpret)  
plt.title("Simple Return")  
plt.xlabel("Index")  
plt.ylabel("Simple Return")  
plt.grid()
plt.show()
```
